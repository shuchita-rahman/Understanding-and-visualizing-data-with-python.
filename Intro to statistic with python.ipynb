### Method of Statistics.

Statistics Method is divided into two categories.

**Descriptive Statistics**: It try to describe find out the feature of the data and then summaries the data by raw observation. Ex: mean, standard deviation.

**Inferential Statistics**: It try to study from a sample of data to give the answer. Inferential Statistics framework is under probability theory.

# Types of Data.
The main intention of data visualization is to find out a summary of the data. Our data summer will be different depending on the data type. Suppose let’s see our data titanic.  Here we easily can see that data all is not numeric. Here **Age**, Fair is numerical value but **Sex**, **Embarked** is not numerical value. If we want to know the average age of the passenger we easily find out by adding the Age column. But If we can’t do the same thing for **Sex**, or **Embarked** column. So for different types of data we need to come up with approach. 
 There are two types of variables:

### Quantitative Data:   
   > Data which deal with number or which we can measure is quantitative data. Ex: Age, Height, Salary.

**Discrete Data:** Just like the name Discrete data are distinct or individual value. We can count Discrete data but we can’t measure it. Discrete data are integer value. Ex: Suppose How many children do you have. The answer can be 0,1 3, 5. But you can’t have 2.5 children. It has to distinct value. 
In titanic database **SibSp**(# of sibling/ spouses), **Parch** (#of children/ parents) are Discrete Quantitative data.


```python
df.loc[:5,["SibSp", "Parch"] ]
```

 **Continuous Data:** Continuous data are those numerical data which can be measured but can not be counted. This type of data can be divided or reduced.Ex : Height, Weight. 
In titanic database **Age**, **Fare** are Continuous Quantitative Data.

```python
df.loc[:5,["Age", "Fare"] ]
```

### Qualitative/ Categorical Data:
   > Qualitative/ Categorical variable deal with characteristic of the data. It can’t be measured by a metric. Ex: Gender, Hair color, Language etc.
    
**Nominal Data:** When a data is labeled or assigned is called nominal data. Nominal doesn’t have ordered so if we change the order it doesn’t change the data meaning. Ex: Suppose tell us the list of  natural hair color? Blonde, Black, Red etc. 
In titanic database **Survived**, **Embarked**, **Sex** are Nominal Qualitative Data.
​

```python
df.loc[:5,["Survived", "Embarked", "Sex"] ]
```

**Ordinal Data:** Ordinal data are those which represent a distinct value but also maintain an order. Ex: Tshirt size Small, Medium, Large.   
In titanic data base **Pclass**(ticket class, 1 = 1st, 2 = 2nd and 3 = 3rd )

```python
df.loc[:5,["Pclass"] ]
```

To gather information about column name and data types we will use the following function.Let’s say we want to see specific column. We can do it using this three way.
*     .loc()
*     .iloc()
*     .ix()

### .loc()
.loc() is a label based searching. It takes 2 operator which is separated by “,”. The first one indicates row and the second one indicates column. Column name doesn’t need to be in order.


```python
# return all rows for Name column
df.loc[:, "Name"]
```

```python
#return first ten rows for Name, Survived and PassengerId columns
df.loc[:3, ["Name", "Survived", "PassengerId"]]
```

```python
#return 10 to 20 rows for Name, Survived and PassergerId columns
df.loc[20:25, ["Name", "Survived", "PassengerId"]]
```

### .iloc()
.iloc() is a integer based slicing. Like .loc() it take input two operator row and column. Difference between .loc() is it can take column number as input.

```python
#show all rows and column
df.iloc[:,:]
```

```python
df.iloc[1:4, 1:4]
```

```python
# show Name , Survived and PassegerId column
df.iloc[:5, [3,1,2]]
```
